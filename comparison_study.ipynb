## Comments:

Note that we are testing a series of samples to decide whether it is drawn from a finite-state Markov chain with a "given" stationary distribution. By "given" we mean such a stationary distribution can be estimated using a long nominal historical sample path of the same Markov chain.

### One possible "simplification" of the asymptotic distribution of the test statistic (based on the Taylor expansion that we have derived and a method mentioned in [Imhof J. P. (1961)]):

$$2n D( \boldsymbol{\Gamma}_n \| \boldsymbol{\pi} ) \approx
  {\textbf{U}_n^{\prime}}{\nabla ^2}h( {\boldsymbol{\pi}} )\textbf{U}_n \xrightarrow[{n \to \infty }]{\textit{d}} \sum\nolimits_{k = 1}^{{N^2}} {{\rho _k}\chi _{1k}^2},  \qquad (1)
$$
where $\rho_k, \, k = 1, \ldots, N^2$, are the eigenvalues of the matrix ${\nabla ^2}h( {\boldsymbol{\pi}} )\boldsymbol{\Sigma}$, and $\chi _{1k}^2, \, k = 1, \ldots, N^2$, are $N^2$ independent $\chi^2$-variables with 1 degree of freedom. 

Given a target false alarm rate $\beta_n$, we can derive a corresponding threshold estimator $\bar{\eta}_n^{{wc}}$ based on (1). However, this still can not avoid simulation. It is hard to conclude any advantage of this "simplification" over what we do in the original paper. We can point out this fact: calculating the eigenvalues of ${\nabla ^2}h( {\boldsymbol{\pi}} )\boldsymbol{\Sigma}$ makes the final estimator of the threshold numerically not as stable, compared to the algorithms that we apply in the paper. Another point we should make is that, in [page 30, Billingsley (1961)], a slightly different statistic is considered and therefore an even simpler asymptotic distribution can be derived correspondingly. Some other authors (e.g., [Menéndez, M. L., et al. (1997)]) also considered similar but different statistics.

Overall comments: Our method is novel considering the specific statistic that we use.
